Lets find the top 10 most frequent Categories of Variable X2 

```python
data.X2.value_counts() 
```

```python
data.X2.value_counts().head(20) 
```

Lets Make a List with the Most frequent categories of Variable 

```python
top_10 = [x for x in data.X2.value_counts().sort_values(ascending = False).head(10).index]  
```

```python
top_10 
```

Now we make 10 binary variables 

```python
for label in top_10:
    data[label] = np.where(data['X2']==label,1,0) 
```

```python
data[['X2']+top_10].head(40) 
```

Get Whole Set of Dummy Variables, for all the Categorical Variables 

```python
def one_hot_top_x(df, variable, top_x_labels):
    # function to create the Dummy variable for the most frequent labels
    # We Can Vary the Number of most frequent labels that we encode
    for label in top_x_labels:
        df[variable+'='+label] = np.where(data[variable]==label,1,0)
        
#Read the Data again 
data = pd.read_csv('F://Data_Set/mercedesbenz.csv',usecols=['X1','X2','X3','X4','X5','X6']) 

# encode X2 in 10 most frequent categories
one_hot_top_x(data,'X2',top_10)
data.head(10) 
```
